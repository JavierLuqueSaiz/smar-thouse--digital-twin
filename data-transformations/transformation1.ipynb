```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data1.csv")
```

```python
df.drop(df.columns[0],axis=1,inplace=True)
```

```python
df.columns
```

```python
n = 21
label = df.columns[n]
plt.hist(df.iloc[:,n])
plt.title(label)
plt.xticks(rotation=45, ha='right')
plt.show()
```

There is an unnecessary amount of possible values

```python
# Renaming Summary variable
# Clear, Cloudy, Rain, Snow, Wind
df["summary"] = df["summary"].replace("Partly Cloudy", "Cloudy")
df["summary"] = df["summary"].replace("Light Rain", "Rain")
df["summary"] = df["summary"].replace("Drizzle", "Rain")
df["summary"] = df["summary"].replace("Overcast", "Cloudy")
df["summary"] = df["summary"].replace("Mostly Cloudy", "Cloudy")
df["summary"] = df["summary"].replace("Light Snow", "Snow")
df["summary"] = df["summary"].replace("Flurries", "Other")
df["summary"] = df["summary"].replace("Breezy", "Other")
df["summary"] = df["summary"].replace("Breezy and Partly Cloudy", "Other")
df["summary"] = df["summary"].replace("Foggy", "Other")
df["summary"] = df["summary"].replace("Rain and Breezy", "Rain")
df["summary"] = df["summary"].replace("Heavy Snow", "Snow")
df["summary"] = df["summary"].replace("Flurries and Breezy", "Snow")
df["summary"] = df["summary"].replace("Breezy and Mostly Cloudy", "Cloudy")
df["summary"] = df["summary"].replace("Dry", "Clear")
```

```python
n = 21
label = df.columns[n]
counts = df.iloc[:, n].value_counts()
plt.bar(counts.index, counts.values)
plt.title(label)
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
df.to_csv("data2.csv")
```
