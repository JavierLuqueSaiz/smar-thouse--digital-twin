```python
df_features = df_Xy[['Mes','Dia','Hora',
       'temperature',
       'humidity', 'visibility', 'apparentTemperature', 'pressure',
       'windSpeed', 'cloudCover', 'windBearing', 'precipIntensity', 'dewPoint',
       'precipProbability', 'Furnace', 'Kitchen', 'Dishwasher..kW.ON',
       'FurnaceON', 'Home.office..kW.ON', 'Wine.cellar..kW.ON',
       'Garage.door..kW.ON', 'KitchenON', 'Barn..kW.ON', 'Well..kW.ON',
       'Microwave..kW.ON', 'Living.room..kW.ON', 'Mes_2', 'Mes_3', 'Mes_4',
       'Mes_5', 'Mes_6', 'Mes_7', 'Mes_8', 'Mes_9', 'Mes_10', 'Mes_11',
       'Mes_12', 'Hora_1', 'Hora_2', 'Hora_3', 'Hora_4', 'Hora_5', 'Hora_6',
       'Hora_7', 'Hora_8', 'Hora_9', 'Hora_10', 'Hora_11', 'Hora_12',
       'Hora_13', 'Hora_14', 'Hora_15', 'Hora_16', 'Hora_17', 'Hora_18',
       'Hora_19', 'Hora_20', 'Hora_21', 'Hora_22', 'Hora_23', 'summary_Cloudy',
       'summary_Other', 'summary_Rain', 'summary_Snow']]
```

```python
df_features = df_features.groupby(by=["Mes","Dia","Hora"]).mean().reset_index()
dates = df_features[["Mes","Dia","Hora"]]
df_features = df_features.drop(["Mes","Dia","Hora"], axis=1)
```

```python
df_objetivos = df_Xy[['Mes','Dia','Hora',
       'gen..kW.', 'House.overall..kW.', 'Dishwasher..kW.', 'Home.office..kW.',
       'Fridge..kW.', 'Wine.cellar..kW.', 'Garage.door..kW.', 'Barn..kW.',
       'Well..kW.', 'Microwave..kW.', 'Living.room..kW.']]
```

```python
df_objetivos = df_objetivos.groupby(by=["Mes","Dia","Hora"]).sum().reset_index().drop(["Mes","Dia","Hora"], axis=1)
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_features)

with open('features.pkl', 'wb') as f:
    pickle.dump(scaled_data, f)

with open('objetivos.pkl', 'wb') as f:
    pickle.dump(df_objetivos, f)

with open('dates.pkl', 'wb') as f:
    pickle.dump(dates, f)
```
